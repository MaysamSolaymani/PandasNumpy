```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/zubairamuti/netflix-movies-and-tv-shows-dataset/croissant/download') # or "./Dataset/netflix_titles.csv.csv"

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
record_set_df.head()
```

```python
df = record_set_df
```

```python
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)
record_set_df.head()
```

```python
usdf = df.where((df['netflix_titles.csv.csv/type']==b'Movie') & (df['netflix_titles.csv.csv/country']==b'United States')).dropna()
```

```python
usdf['netflix_titles.csv.csv/listed_in'] = usdf['netflix_titles.csv.csv/listed_in'].apply(lambda v: v.decode() if isinstance(v, bytes) else str(v))
df_filtered = usdf[~usdf['netflix_titles.csv.csv/listed_in'].str.contains('Horror', case=False, na=False)]
df_filtered.count()
```

```python
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('Comedies', case=False, na=False)]
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('Music', case=False, na=False)]
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('Stand-Up Comedy', case=False, na=False)]
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('Children & Family Movies', case=False, na=False)]
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('LGBTQ', case=False, na=False)]
df_filtered = df_filtered[df_filtered['netflix_titles.csv.csv/release_year'] > 1980]
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/listed_in'].str.contains('Documentaries', case=False, na=False)]
```

```python
df_filtered['netflix_titles.csv.csv/description'] = df_filtered['netflix_titles.csv.csv/description'].apply(lambda v: v.decode() if isinstance(v, bytes) else str(v))
df_filtered = df_filtered[~df_filtered['netflix_titles.csv.csv/description'].str.contains('Animation', case=False, na=False)]
df_filtered['time_str'] = df_filtered['netflix_titles.csv.csv/duration'].apply(lambda x: x.decode('utf-8'))
df_filtered['time_int'] = df_filtered['time_str'].str.extract(r'(\d+)').astype(int)
df_sorted = df_filtered.sort_values('time_int', ascending=False)
df_sorted
```

```python
df_filtered['netflix_titles.csv.csv/cast'] = df_filtered['netflix_titles.csv.csv/cast'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```
