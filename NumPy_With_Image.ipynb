```python
#NumPy (Numerical Python) is a library for handling arrays, matrices, and performing high-performance numerical computations.
# It’s way faster than using plain Python lists for numbers because it’s written in C under the hood.
import numpy as np
# skimage (or scikit-image) is a Python library built on top of NumPy and SciPy, specifically for image processing and computer vision
from skimage import io
# creating static, animated, and interactive visualizations
from matplotlib import pyplot as plt
```

```python
photo = io.imread("./Photos/Levels.png")
plt.imshow(photo)
```

```python
flipped = photo[:, ::-1] # flip horizontally np.flip(photo, axis=1)
plt.imshow(flipped)
```

```python
photo_masked = np.where(photo>100, 255, photo)
plt.imshow(photo_masked)
```
