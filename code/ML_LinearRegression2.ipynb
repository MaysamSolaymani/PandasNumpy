```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
yeas_of_experience = np.array([0,1,2,3,4,6,8,10]).reshape(-1,1)
salary = np.array([30000,40000,55000,60000,70000,80000,85000,87000])
plt.plot(yeas_of_experience, salary,'*')
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=4)
model.fit(yeas_of_experience, salary)
plt.plot(yeas_of_experience,salary,'*')
plt.plot(yeas_of_experience, model.predict(yeas_of_experience))
print(model.max_depth)
```
