```python
import mlcroissant as mlc
import pandas as pd

# Fetch the Croissant JSON-LD
croissant_dataset = mlc.Dataset('https://www.kaggle.com/datasets/zubairamuti/netflix-movies-and-tv-shows-dataset/croissant/download') # or "./Dataset/netflix_titles.csv.csv"

# Check what record sets are in the dataset
record_sets = croissant_dataset.metadata.record_sets
print(record_sets)

# Fetch the records and put them in a DataFrame
record_set_df = pd.DataFrame(croissant_dataset.records(record_set=record_sets[0].uuid))
```

```python
df = record_set_df.copy()
df.head(7)
```

```python
df = df.rename(columns={"netflix_titles.csv.csv/show_id": "show_id" , "netflix_titles.csv.csv/type" : "type", "netflix_titles.csv.csv/title":"title",
                        "netflix_titles.csv.csv/director":"director", "netflix_titles.csv.csv/cast":"cast", "netflix_titles.csv.csv/country":"country",
                        "netflix_titles.csv.csv/date_added":"date_added", "netflix_titles.csv.csv/release_year":"release_year", "netflix_titles.csv.csv/rating":"rating",
                        "netflix_titles.csv.csv/duration":"duration", "netflix_titles.csv.csv/listed_in":"listed_in", "netflix_titles.csv.csv/description":"description"})
df
```

```python
usdf = df.where((df['type']==b'Movie') & (df['country']==b'United States')).dropna()
usdf['listed_in'] = usdf['listed_in'].apply(lambda v: v.decode() if isinstance(v, bytes) else str(v))
exclude = ['Horror', 'Comedies', 'Music', 'Stand-Up Comedy', 'Children & Family Movies', 'LGBTQ', 'Documentaries']
df_filtered = usdf[~usdf['listed_in'].str.contains('|'.join(exclude), case=False, na=False)]
df_filtered = df_filtered[df_filtered['release_year'] > 1980]
df_filtered
```

```python
df_filtered['time_str'] = df_filtered['duration'].apply(lambda x: x.decode('utf-8'))
df_filtered['time_int'] = df_filtered['time_str'].str.extract(r'(\d+)').astype(int)
df_sorted = df_filtered.sort_values('time_int', ascending=False)
df_sorted
```

```python
df_filtered['cast'] = df_filtered['cast'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
df_filtered['director'] = df_filtered['director'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
DavidFincher = df_filtered[df_filtered['director'].str.contains('David Fincher', case=False, na=False)]
DavidFincher
```

```python
StevenSpielberg = df_filtered[df_filtered['director'].str.contains('Steven Spielberg', case=False, na=False)]
StevenSpielberg
```
